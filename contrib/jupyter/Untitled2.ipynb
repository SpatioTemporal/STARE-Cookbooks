```python
def mk_tri0(m,n,L):
    x0=m/L; x1=(m+1)/L
    y0=n/L; y1=(n+1)/L
    return numpy.array([x0+y0*1j,x1+y0*1j,x0+y1*1j,x0+y0*1j])

def mk_tri1(m,n,L):
    x0=m/L; x1=(m+1)/L
    y0=n/L; y1=(n+1)/L
    return numpy.array([x1+y0*1j,x1+y1*1j,x0+y1*1j,x1+y0*1j])
```
